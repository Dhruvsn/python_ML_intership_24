```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

def calc_grade(perc):
    perc = float(perc) 
    if perc <= 100 and perc >= 80:
        return 'A'
    elif perc < 80 and perc >= 60:
        return 'B'
    elif perc < 60 and perc >= 50:
        return 'C'
    elif perc < 50 and perc >= 40:
        return 'D'
    elif perc < 40 and perc >= 30:
        return 'E'
    elif perc < 30 and perc >= 0:
        return 'F'
    else:
        return 'x'

# Read the CSV file
cd = pd.read_csv('student_marks.csv')

# Add 'Total' and 'Percentage' columns
cd['Total'] = cd[['English', 'Hindi', 'Maths', 'Science', 'SST']].sum(axis=1)
cd['Percentage'] = (cd['Total'] / 5).apply(lambda x: f"{x:.2f}")

# Calculate the 'Grade' column
cd['Grade'] = cd['Percentage'].apply(calc_grade)

# Use the full DataFrame or the first few rows as needed
temp = cd.head()

# Define subjects and colors
sub = ['English', 'Hindi', 'Maths', 'Science', 'SST']
col = ['r', 'b', 'g', 'y', 'm']

# Plotting each subject
# for i in range(5):
#     plt.figure(figsize=(10, 6))
#     plt.plot(temp['Name'], temp[sub[i]], marker='o', label=sub[i], color=col[i])
#     plt.title(f"Marks in {sub[i]}")
#     plt.xlabel('Name')
#     plt.ylabel('Marks')
#     plt.legend()
#     plt.grid(True)
#     # plt.xticks(rotation=45)  # Rotate x-axis labels if needed
#     plt.show()
```
