```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
```

```python
news = pd.read_csv('titanic.csv')
news

peopleLive = news[['sex','survived']]
peopleLive = peopleLive.groupby(['sex','survived']).size().unstack(fill_value=0)
peopleLive.plot(kind='bar',stacked=False)

```

```python
plt.figure(figsize=(20,8))
sns.barplot(data=news['survived'],linewidth=1.5,label="survived")
sns.barplot(data=news['age'],linewidth=1.5,label='age')

plt.title('Female Vs male survived')
plt.ylabel('survived')
plt.xlabel('age')
plt.show()
```
